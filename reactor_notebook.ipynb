```python
import tkinter as tk

root = tk.Tk()
root.geometry("600x400")
enrichment_var = tk.DoubleVar(value=0.72)

def submit():
    print(enrichment_var.get())

enrichment_entry = tk.Entry(root, textvariable=enrichment_var)
enrichment_entry.pack()

submit_button = tk.Button(root, text="sbubmit", command=submit)
submit_button.pack()
    
root.mainloop()
```
